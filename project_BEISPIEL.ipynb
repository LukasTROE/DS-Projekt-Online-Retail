### 2.4 Feature Engineering - Investoren-relevante Metriken erstellen

```python
# Wichtige Investoren-Metriken berechnen
print("üîß Feature Engineering: Erstelle Investment-Metriken...")
print("="*80)

# 1. ROI (Return on Investment) - Valuation / Funding
df['ROI_Multiple'] = (df['Valuation (M USD)'] / df['Funding Amount (M USD)']).round(2)

# 2. Revenue Efficiency - Revenue pro Million Funding
df['Revenue_per_Funding'] = (df['Revenue (M USD)'] / df['Funding Amount (M USD)']).round(2)

# 3. Valuation to Revenue Ratio (√§hnlich wie P/E Ratio)
df['Valuation_to_Revenue'] = (df['Valuation (M USD)'] / df['Revenue (M USD)']).replace([np.inf, -np.inf], np.nan).round(2)

# 4. Revenue per Employee (Produktivit√§t)
df['Revenue_per_Employee'] = ((df['Revenue (M USD)'] * 1_000_000) / df['Employees']).round(0)

# 5. Funding per Employee
df['Funding_per_Employee'] = ((df['Funding Amount (M USD)'] * 1_000_000) / df['Employees']).round(0)

# 6. Company Age (Alter seit Gr√ºndung)
current_year = 2025
df['Company_Age'] = current_year - df['Year Founded']

# 7. Exit Success (1 wenn IPO oder Acquired, 0 wenn Private)
df['Successful_Exit'] = df['Exit Status'].apply(lambda x: 1 if x in ['IPO', 'Acquired'] else 0)

# 8. Funding Stage Kategorisierung
def categorize_funding_stage(rounds):
    if rounds <= 2:
        return 'Early Stage'
    elif rounds <= 3:
        return 'Growth Stage'
    else:
        return 'Late Stage'

df['Funding_Stage'] = df['Funding Rounds'].apply(categorize_funding_stage)

print("‚úÖ Feature Engineering abgeschlossen!")
print(f"\nüìä Neue Features erstellt:")
print("   ‚Ä¢ ROI_Multiple: Valuation / Funding Ratio")
print("   ‚Ä¢ Revenue_per_Funding: Umsatzeffizienz")
print("   ‚Ä¢ Valuation_to_Revenue: Bewertungsmultiplikator")
print("   ‚Ä¢ Revenue_per_Employee: Produktivit√§t")
print("   ‚Ä¢ Funding_per_Employee: Kapitalintensit√§t")
print("   ‚Ä¢ Company_Age: Alter des Startups")
print("   ‚Ä¢ Successful_Exit: Exit-Erfolg (bin√§r)")
print("   ‚Ä¢ Funding_Stage: Entwicklungsstufe")

print("\n" + "="*80)
print("Beispiel der neuen Features:")
df[['Startup Name', 'Industry', 'ROI_Multiple', 'Revenue_per_Funding', 
    'Company_Age', 'Funding_Stage', 'Successful_Exit']].head(10)
```

### 3.5 Risk-Return Analyse - Portfolio-Optimierung

## 4. Machine Learning - Predictive Analytics f√ºr Investoren

### 4.1 Profitabilit√§ts-Vorhersage Modell

```python
# Machine Learning: Profitabilit√§ts-Vorhersage
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import LabelEncoder

print("ü§ñ MACHINE LEARNING: Profitabilit√§ts-Vorhersage")
print("="*100)

# Feature Engineering f√ºr ML
df_ml = df.copy()

# Label Encoding f√ºr kategoriale Variablen
le_industry = LabelEncoder()
le_region = LabelEncoder()
le_exit = LabelEncoder()

df_ml['Industry_encoded'] = le_industry.fit_transform(df_ml['Industry'])
df_ml['Region_encoded'] = le_region.fit_transform(df_ml['Region'])
df_ml['Exit_Status_encoded'] = le_exit.fit_transform(df_ml['Exit Status'])

# Features ausw√§hlen (ohne Target-Leakage)
features = ['Funding Rounds', 'Funding Amount (M USD)', 'Valuation (M USD)', 
            'Revenue (M USD)', 'Employees', 'Market Share (%)', 'Company_Age',
            'Industry_encoded', 'Region_encoded', 'ROI_Multiple']

X = df_ml[features].fillna(df_ml[features].median())
y = df_ml['Profitable']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)

# Vorhersagen
y_pred = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]

# Evaluation
print("\nüìä MODEL PERFORMANCE:")
print("="*100)
print(classification_report(y_test, y_pred, target_names=['Nicht profitabel', 'Profitabel']))

print(f"\nüéØ ROC-AUC Score: {roc_auc_score(y_test, y_pred_proba):.3f}")

# Feature Importance
print("\n" + "="*100)
print("üîç WICHTIGSTE FEATURES F√úR PROFITABILIT√ÑT:")
print("="*100)

feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

# Feature Namen zur√ºck √ºbersetzen
feature_importance['Feature'] = feature_importance['Feature'].replace({
    'Industry_encoded': 'Industry',
    'Region_encoded': 'Region',
    'Exit_Status_encoded': 'Exit Status'
})

print(feature_importance.head(10))

print("\nüí° INVESTMENT INSIGHT:")
print(f"Die wichtigsten Faktoren f√ºr Profitabilit√§t sind:")
for i, row in feature_importance.head(5).iterrows():
    print(f"  {i+1}. {row['Feature']}: {row['Importance']:.3f}")
```

## 5. Investment Recommendations - Handlungsempfehlungen

### 5.1 Top Investment Opportunities nach Branche

```python
# Investment Scoring System
print("üéØ INVESTMENT SCORING & EMPFEHLUNGEN")
print("="*100)

# Scoring-Kriterien f√ºr Investment-Attraktivit√§t
industry_scores = industry_analysis.copy()

# Normalisiere alle Metriken auf 0-100 Skala
industry_scores['ROI_Score'] = ((industry_scores['√ò ROI Multiple'] - industry_scores['√ò ROI Multiple'].min()) / 
                                 (industry_scores['√ò ROI Multiple'].max() - industry_scores['√ò ROI Multiple'].min()) * 100)

industry_scores['Exit_Score'] = industry_scores['Exit Success (%)']
industry_scores['Profit_Score'] = industry_scores['Profitabilit√§t (%)']

# Market Share als Indikator f√ºr etablierte Branche
industry_scores['Market_Score'] = ((industry_scores['√ò Market Share (%)'] - industry_scores['√ò Market Share (%)'].min()) / 
                                    (industry_scores['√ò Market Share (%)'].max() - industry_scores['√ò Market Share (%)'].min()) * 100)

# Gesamtscore (gewichteter Durchschnitt)
weights = {
    'ROI_Score': 0.35,      # Wichtigster Faktor
    'Exit_Score': 0.30,     # Exit-Potential sehr wichtig
    'Profit_Score': 0.25,   # Profitabilit√§t wichtig
    'Market_Score': 0.10    # Market Share weniger gewichtet
}

industry_scores['Investment_Score'] = (
    industry_scores['ROI_Score'] * weights['ROI_Score'] +
    industry_scores['Exit_Score'] * weights['Exit_Score'] +
    industry_scores['Profit_Score'] * weights['Profit_Score'] +
    industry_scores['Market_Score'] * weights['Market_Score']
).round(1)

# Nach Investment Score sortieren
investment_ranking = industry_scores[['Investment_Score', 'ROI_Score', 'Exit_Score', 
                                      'Profit_Score', '√ò Funding (M$)', '√ò ROI Multiple']].sort_values(
    'Investment_Score', ascending=False
)

print("\nüìä BRANCHEN-RANKING (Investment Attractiveness Score):")
print("="*100)
print(investment_ranking)

# Top 3 Empfehlungen
print("\n" + "="*100)
print("üèÜ TOP 3 INVESTMENT-EMPFEHLUNGEN:")
print("="*100)

for i, (industry, row) in enumerate(investment_ranking.head(3).iterrows(), 1):
    print(f"\n{i}. {industry}")
    print(f"   Investment Score: {row['Investment_Score']:.1f}/100")
    print(f"   ‚Ä¢ Durchschnittlicher ROI: {row['√ò ROI Multiple']:.2f}x")
    print(f"   ‚Ä¢ Exit Success Rate: {row['Exit_Score']:.1f}%")
    print(f"   ‚Ä¢ Profitabilit√§ts-Rate: {row['Profit_Score']:.1f}%")
    print(f"   ‚Ä¢ Typisches Investment: ${row['√ò Funding (M$)']:.1f}M")
    
    # Empfehlung basierend auf Score-Profil
    if row['ROI_Score'] > 80:
        print(f"   ‚úÖ Empfehlung: HOHE PRIORIT√ÑT - Exzellentes ROI-Potential")
    elif row['Investment_Score'] > 60:
        print(f"   ‚úÖ Empfehlung: MITTEL-HOCH - Solide Investment-Option")
    else:
        print(f"   ‚ö†Ô∏è Empfehlung: MIT VORSICHT - Gr√ºndliche Due Diligence erforderlich")
```
